```python
import json 
import pandas as pd
import numpy as np
```

```python
all_data = json.load(open("../year-all.json"))
df = pd.read_csv("co-author-aff.csv")
```

```python
df.Affiliations.fillna("Undefined" ,inplace = True)
```

```python
df.isna().sum()
```

```python
coaffMap = {k:v for k,v in zip(df.Authors.values.tolist(), df.Affiliations.values.tolist())}
```

```python
coaffMap
```

```python
for node in all_data["nodes"]:
    try:
        node["aff"] = coaffMap[node["id"]]
    except:
        node["aff"] = "Not Available"
```

```python
with open("year-all-aff.json", "w") as f:
    json.dump(all_data, f, indent = 4)
```
